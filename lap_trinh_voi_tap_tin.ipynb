```python
import random as r
def cau_5():
   x = r.choices(range(-1000,1001),k=1000)
    print('danh s√°ch: ',x)
    ten_tap_tin =input("Nhap ten tap tin: ")
    f=open(ten_tap_tin,"w")
    a=[]
    for i in range(10):
        a.append(i)
    for i in range(100):
        for j in range(10):
            a[j]=str(x[i*10+j])
        f.write(','.join(a)+"\n")
    f.close()
    with open(ten_tap_tin,"r+") as f:
        tep=f.read()
        tep=tep.replace(",", "  ")
        print(tep)
cau_5()
```
